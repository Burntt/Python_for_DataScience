```python
import pandas as pd
import numpy as np

from sklearn import model_selection
```

## Parameters search

```python
titanic_df = pd.read_csv('titanic.csv')
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(
    titanic_df.drop('Survived', axis=1),
    titanic_df['Survived'])
```

```python
X_train
```

```python
%store X_train
```

#### for classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classifier = RandomForestClassifier(n_estimators=300, random_state=0)
```

```python
grid_param = {
    'n_estimators': [100, 300, 500, 800, 1000],
    'criterion': ['gini', 'entropy'],
    'bootstrap': [True, False],
    'max_depth': [1,2,3,6, None]
}
```

```python
from sklearn import model_selection
```

### Grid Search

```python
grid_search_classifier = model_selection.GridSearchCV(estimator=classifier,
                     param_grid=grid_param,
                     scoring='accuracy',
                     cv=5,
                     n_jobs=-1)
grid_search_classifier
```

```python
%%time
grid_search_classifier.fit(X_train[ ['Pclass', 'SibSp','Fare' ] ], y_train)
```

```python
best_parameters = grid_search_classifier.best_params_
print(best_parameters)
```

```python
grid_search_classifier.best_score_
```

```python
grid_search_classifier.cv_results_
```

```python
len(grid_search_classifier.cv_results_['params'])
```

### Random search

```python
random_search_classifier = model_selection.RandomizedSearchCV(estimator=classifier,
                     param_distributions = grid_param,
                     n_iter=30,
                     scoring='accuracy',
                     cv=5,
                     n_jobs=-1)
```

```python
%%time
random_search_classifier.fit(X_train[ ['Pclass', 'SibSp','Fare' ] ], y_train)
```

```python
best_parameters = random_search_classifier.best_params_
print(best_parameters)
```

```python
random_search_classifier.best_score_
```

```python
grid_search_classifier.cv_results_
```

```python
random_search_classifier_2 = model_selection.RandomizedSearchCV(estimator=classifier,
                     param_distributions = param_distribution,
                     n_iter=30,
                     scoring='accuracy',
                     cv=5,
                     n_jobs=-1)
```

#### for pipeline

```python
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV


pca = PCA()
logistic = LogisticRegression(max_iter=10000, tol=0.1)
pipe = Pipeline(steps=[('pca', pca), ('logistic', logistic)])
```

```python
X_train[ ['Pclass', 'SibSp','Fare' ] ]
```

```python
range(2,10)
```

```python
pipe_param_distribution = {
    'pca__n_components': range(2,11),
    'pca__svd_solver':['auto','randomized'],
    'logistic__penalty':['l2','l1'],
    'logistic__max_iter': range(700,1201)
}
```

```python
random_search_pipeline = model_selection.RandomizedSearchCV(estimator=pipe,
                     param_distributions = pipe_param_distribution,
                     n_iter=10,
                     scoring='accuracy',
                     cv=5,
                     n_jobs=-1)
```

```python
%%time
random_search_pipeline.fit(X_train[ ['Pclass', 'SibSp','Fare' ] ], y_train)
```

```python
best_parameters = random_search_pipeline.best_params_
print(best_parameters)
```

```python
random_search_pipeline.best_score_
```

#### Exercise 1: apply hyperparameter search to pokemon dataset optimization

```python
pokemon_df = pd.read_csv('Pokemon.csv')
pokemon_df['HighStage'] = pokemon_df['Stage'] > 2
pokemon_df['Bin_HighStage'] = pokemon_df['HighStage'].apply(lambda x: 1 if x == True else 0)
```

```python
# iterating the columns 
pok_feature_names = []
for col in pokemon_df.columns: 
    pok_feature_names.append(str(col))
pok_feature_names[1:-1]

pokemon_df.columns
```

```python
pokemon_df = pokemon_df.drop(pok_feature_names[1:4], axis=1)
pokemon_df = pokemon_df.drop(['Legendary', 'Stage', 'HighStage'], axis=1)
pokemon_df
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(
    pokemon_df.drop('Bin_HighStage', axis=1),
    pokemon_df['Bin_HighStage'])
```

```python
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from scipy import stats


pca = PCA()
logistic = LogisticRegression(max_iter=10000, tol=0.1)
pipe = Pipeline(steps=[('pca', pca), ('logistic', logistic)])
```

```python
param_distribution = {
    'n_estimators': stats.poisson(300),
    'criterion': ['gini', 'entropy'],
    'bootstrap': [True, False],
    'max_depth': stats.binom(10,0.3)
}
```

```python
random_search_classifier_2 = model_selection.RandomizedSearchCV(estimator=classifier,
                     param_distributions = param_distribution,
                     n_iter=30,
                     scoring='accuracy',
                     cv=5,
                     n_jobs=-1)
```

```python
feat_names = \
['Total',
 'HP',
 'Attack',
 'Defense',
 'Sp_Atk',
 'Sp_Def',
 'Speed']
```

```python
%%time
random_search_classifier_2.fit(X_train[feat_names],y_train)
```
