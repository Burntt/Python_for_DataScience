#### General imports

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns

from bokeh.io import output_file, output_notebook
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
from bokeh.layouts import row, column, gridplot
from bokeh.models.widgets import Tabs, Panel
```

```python
pokemon_df = pd.read_csv('https://raw.githubusercontent.com/ObozniyMaksym/HS_data/main/Pokemon_m.csv')
```

```python
pokemon_source = ColumnDataSource(pokemon_df)
output_notebook()
```

### Scatterplots

#### Pandas

```python
pokemon_df.plot(kind = 'scatter',
                x = 'Attack',
               y = 'Defense')
```

#### Matplotlib

```python
plt.scatter('Attack', 'Defense', data = pokemon_df)
```

#### Seaborn

```python
sns.scatterplot(x='Attack', y='Defense', data=pokemon_df)
```

#### Bokeh

```python
fig = figure()
fig.circle(x='Attack', y='Defense',
         source=pokemon_source,
         size=10, color='blue')


fig.title.text = 'Pokemon dataset'
fig.xaxis.axis_label = 'Attack'
fig.yaxis.axis_label = 'Defense'

show(fig)
```
