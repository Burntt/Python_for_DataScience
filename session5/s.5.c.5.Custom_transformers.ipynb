#### fullscale partial transformer

```python
from sklearn.base import TransformerMixin
```

```python
class FunctionFeaturizer(TransformerMixin):
    def __init__(self, *featurizers, seventeen=17):
        self.seventeen = seventeen

    def fit(self, X, y=None):
        self.seventeen = - np.mean(X)
        return self

    def transform(self, X):
        #Do transformations
        return X+self.seventeen
```

```python
our_transformer = FunctionFeaturizer(seventeen=19)
X = np.array([[0, 1], [2, 3]])
our_transformer.transform(X)
```
