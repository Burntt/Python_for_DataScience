#### Exercise 2.3: create pipeline from scaler, normalizer, polynomial features, PCA and Linear regression, split sample to train and test, fit the pipeline on train and then predict it on test

##### Preprocess

##### Create all pipelines

```python
from sklearn.preprocessing import *
from sklearn.decomposition import*
from sklearn.neighbors import*
```

```python
%store -r pokemon_df
pokemon_df
```

```python
features = ['Total', 'Attack', 'Defense', 'Sp_Def', 'Speed', 'Stage']
target = 'Sp_Atk'
```

```python
pipe = Pipeline([('scaler', StandardScaler()), \
                 ('normalizer', Normalizer()), \
                 ('polynomial features', PolynomialFeatures()), \
                 ('pca', PCA()), \
                 ('regression', LinearRegression())])
```

```python
# First way of manual split

X_train, X_test, y_train, y_test = model_selection.train_test_split(
    pokemon_df[features], 
    pokemon_df[target], 
    test_size=0.2, 
    random_state=2020)
```

```python
# Second way, k-fold split

acc = 0
folds = 5
for train_index,test_index in model_selection.KFold(folds).split(pokemon_df):
    #print(train_index.shape, test_index.shape)
    #print(pokemon_df.loc[train_index,'Sp_Atk'].mean()) 
    
    train_index = train_index[1:]
    test_index = test_index[1:]
    
    X_train = pokemon_df.loc[train_index, features]
    y_train = pokemon_df.loc[train_index, 'Sp_Atk']
    X_test = pokemon_df.loc[test_index, features]
    y_test = pokemon_df.loc[test_index, 'Sp_Atk']
    
    #print(X_test)
    pipeline_estimator = pipe.fit(X = X_train, y = y_train)
    y_pred = pipeline_estimator.predict(X = X_test)
    accuracy = metrics.mean_absolute_error(y_test, y_pred)
    acc = acc + accuracy
    print(accuracy)
    
mean_acc = acc/folds
print(mean_acc)
```
