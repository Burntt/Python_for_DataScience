## Mutual information score

```python
X = np.random.binomial(5,0.9,size=100)

Y = X + 1 * np.random.binomial(1,0.05,size=100)
```

```python

sns.distplot(X)
```

```python
metrics.mutual_info_score(X,X)
```

```python
metrics.mutual_info_score(X,Y)
```

```python
metrics.adjusted_mutual_info_score(X,X)
```

```python
metrics.adjusted_mutual_info_score(X,Y)
```

```python
datapoints = []
for i in range(-100,100):
    alpha = i*0.01
    
    X = np.random.binomial(5,0.5,size=100)

    Y = X + alpha * np.random.binomial(1,0.05,size=100)
    
    datapoints.append( (alpha,  metrics.adjusted_mutual_info_score(X,Y)) ) 
    

mututal_info_df = pd.DataFrame( datapoints, columns =['alpha','mutual_info'] )
mututal_info_df.plot()
```

#### Exercise 2.4: Study the connection between lambda value of puasson variable and mutual information

```python
datapoints = []
for i in range(0, 100):
    X = np.random.binomial(5, 0.5, 100)
    Y = X + np.random.poisson(lam = i / 100, size = 100)
    datapoints.append( (i / 100, metrics.adjusted_mutual_info_score(X, Y)) )

mutual_info_df = pd.DataFrame(datapoints, columns = ['lambda', 'mutual info'])
mutual_info_df.plot()
```
