#### Exercise 1.2: Create basic pipeline for pokemon dataset

```python
pokemon_pipeline = Pipeline(
    [('scale', preprocessing.StandardScaler()),
     ('normalize', preprocessing.Normalizer()),
     ('classifier', LinearRegression())])
```

```python
target = pokemon_df["Total"]
new_pokemon_df = pokemon_df.drop(['Name', "Type_1", "Type_2", "Total", "Legendary"], axis = 1)
#new_pokemon_df
```

```python
fitted_pokemon = pokemon_pipeline.fit(X = new_pokemon_df, y = target)
```

```python
target
```

```python
prediction = fitted_pokemon.predict(new_pokemon_df)
#prediction
```

```python
acc = sklearn.metrics.mean_squared_error(target, prediction)
acc
```

```python
fitted_pipeline = our_pipeline.fit(new_pokemon_df)
new_pokemon_df = fitted_pipeline.transform(new_pokemon_df)
#new_pokemon_df
```
