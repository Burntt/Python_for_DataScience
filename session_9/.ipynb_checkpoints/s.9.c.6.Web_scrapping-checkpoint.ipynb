#### Exercise 2: Try word2vec embeding with gensim

```python
import nltk
from gensim.models import Word2Vec, KeyedVectors
nltk.download('all')
```

```python
text_array_tokenized = [nltk.word_tokenize(title) for title in text_array]
text_array_tokenized
```

```python
model = Word2Vec(text_array_tokenized, min_count =1, size = 32)
model.save("word2vec.model")
```

```python
model = Word2Vec.load("word2vec.model")
model.train([["hello", "world"]], total_examples=1, epochs=1)
```

```python
test = model.most_similar('progressive')
test
```

```python
test1 = model['Computer']
model.most_similar([test1])
```

```python
test2 = model['Computer']
model.most_similar([test2])
```

```python
print(model.predict_output_word(['called','Youth', 'Probability', 'realisation']))
```

```python
from gensim.models import Word2Vec
import numpy as np
from scipy.spatial.distance import cdist
import seaborn as sns
import matplotlib.pyplot as plt

_ = {word: idx for (idx, word) in enumerate(model.wv.index2word)}  # for comfort
similarity = 1 - cdist(model.wv.syn0, model.wv.syn0, metric='cosine')


fig, ax = plt.subplots(figsize=(10,10))         # Sample figsize in inches
sns.heatmap(similarity, xticklabels=model.wv.index2word, yticklabels=model.wv.index2word, annot=True, linewidths=.5, ax=ax)
```
