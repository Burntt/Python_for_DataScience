### Bag of chars

```python
vectorizer = sk_text.CountVectorizer(analyzer='char')
X_train_feat = vectorizer.fit_transform(X_train)
print(X_train_feat.shape)
print(vectorizer.get_feature_names()[:10], '...', vectorizer.get_feature_names()[-10:])
```

```python
len(vectorizer.get_feature_names())
```

```python
character_counts = pd.DataFrame(X_train_feat.todense(), 
                                columns = vectorizer.get_feature_names(),
                               index = X_train.index)
character_counts
```

### Bag of words cleaned

```python
vectorizer = sk_text.CountVectorizer(min_df=10, 
                                     stop_words='english',
                                    token_pattern=r'\b[a-zA-Z]+\b')
X_train_feat = vectorizer.fit_transform(X_train)
```

```python
print(vectorizer.get_feature_names()[:10], '...', vectorizer.get_feature_names()[-10:])
```

```python
X_train_feat.shape
```

```python
X_train_feat.todense()
```

```python
print(X_train_feat.shape)
print(np.max(X_train_feat.toarray()))
```

### Bag of words cleaned with n-grams

```python
vectorizer = sk_text.CountVectorizer(min_df=5, 
                                     stop_words='english',
                                    ngram_range=(1,3))
X_train_feat = vectorizer.fit_transform(X_train)
```

```python
print(vectorizer.get_feature_names()[:10], '...', vectorizer.get_feature_names()[-10:])
```

```python
X_train_feat.shape
```

```python
X_train_feat.todense()
```

```python
print(X_train_feat.shape)
print(np.max(X_train_feat.toarray()))
```
